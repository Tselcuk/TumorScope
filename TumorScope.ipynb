```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np 
from sklearn.linear_model import LogisticRegression 
```

```python
df = pd.read_csv('BrainTumor.csv')
df.head()
```

```python
data = pd.read_csv('BrainTumor.csv', header=0)
yX = data.values.tolist() 
yX = np.array(yX)
yX = yX[:, 1:]
print(yX)
```

```python
y = yX[:100, :1]
y = y.T
y = y[0]
print(y.shape)
```

```python
X = yX[:100, 1:]
print(X.shape)
```

```python
X = np.array(X).astype(float)
y = np.array(y).astype(float)

lr_model = LogisticRegression(max_iter=1500, verbose=1)
lr_model.fit(X, y)
```

```python
print( "score" , lr_model.score(X, y))
```

```python
print("Input the 13 allocated fields to find if tumor is cancerous:")
p = np.zeros(13).reshape(1,13)
for i in range (13): 
    p[0, i]=input(f'Field {i+1} ')
if(lr_model.predict(p)[0] == 0):
    print("Tumour is likely not cancerous")
else:
    print("The tumor may be cancerous")
```
